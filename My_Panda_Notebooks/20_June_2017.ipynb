```python
import numpy as np
import pandas as pd
df = pd.read_csv('/home/greyatom/titanic.csv')
df.isnull().any()
passenger_class = df['Pclass'].value_counts().index
df['Embarked'].isnull()
fares_by_class = {}
for current_class in passenger_class:
    is_current_class = df['Pclass'] == current_class
    current_class_rows = df[is_current_class]
    current_class_fares = current_class_rows['Fare']
    fares_by_class [current_class] = current_class_fares.mean()
    
fares_by_class
```
